```python
# 1. Define the path to your data folder
folder_path = 'CalHABMAP Data/cleaned_hab_data'

# 2. Get a list of all CSV files starting with 'HABs'
file_pattern = os.path.join(folder_path, "HABs*.csv")
all_files = glob.glob(file_pattern)

all_df_list = []

for filename in all_files:
    # Read the individual file
    df = pd.read_csv(filename)
    
    # 3. Extract the raw name (e.g., 'BodegaMarineLab')
    base_name = os.path.basename(filename)
    raw_name = base_name.split('HABs-')[1].split('_')[0]
    
    # 4. Use Regex to insert spaces between CamelCase words
    # This finds a lowercase letter followed by an uppercase letter and adds a space
    pier_name = re.sub(r'([a-z])([A-Z])', r'\1 \2', raw_name)
    
    # Add the cleaned column
    df['station'] = pier_name
    
    all_df_list.append(df)

# 5. Concatenate all dataframes into one
combined_hab_df = pd.concat(all_df_list, axis=0, ignore_index=True)

# Save the result
# combined_hab_df.to_csv('combined_hab_data.csv', index=False)

print("Combined file created successfully with formatted pier names.")
print(len(combined_hab_df['station'].unique()))
combined_hab_df
```

```python
# Keep only essential columns
core_columns = [
    'week_start', 'station', 'sample_date', 'latitude', 'longitude', 'month', 'year',
    'pda', 'temp', 'silicate', 'nitrate', 'avg_chloro',
]
clean_df = combined_hab_df[core_columns].copy()

# Remove all rows with pda = NaN
clean_df = clean_df[combined_hab_df['pda'].notna()].reset_index(drop=True)
clean_df
```

```python
# This creates a table showing how many NaNs each station has for each column
station_nan_check = clean_df.groupby('station').apply(lambda x: x.isnull().sum())
print(station_nan_check[['temp', 'silicate', 'nitrate', 'pda']])
```

```python
# 1. Define the bloom threshold
threshold = 0.1

# 2. Calculate counts per station
bloom_counts = clean_df[clean_df['pda'] > threshold].groupby('station')['pda'].count()
total_samples = clean_df.groupby('station')['pda'].count()

# 3. Combine into a summary table
bloom_report = pd.DataFrame({
    'Bloom Events': bloom_counts,
    'Total Samples': total_samples
}).fillna(0) # Replace NaNs with 0 for stations with no blooms

# 4. Calculate the frequency (Percentage of samples that are toxic)
bloom_report['Bloom Rate (%)'] = (bloom_report['Bloom Events'] / bloom_report['Total Samples'] * 100).round(2)

# Sort by the most toxic stations
print(bloom_report.sort_values(by='Bloom Events', ascending=False))
```

```python
# Cleaning and Imputing based on your station check
for col in ['temp', 'silicate', 'nitrate', 'avg_chloro']:
    # Interpolate small gaps within stations
    clean_df[col] = clean_df.groupby('station')[col].transform(
        lambda x: x.interpolate(limit_direction='both')
    )
    # Fill empty stations (like Trinidad) with the monthly average of the whole coast
    clean_df[col] = clean_df[col].fillna(
        clean_df.groupby('month')[col].transform('median')
    )

# Create Binary Target (Potential Bloom vs No Bloom)
clean_df['potential_bloom'] = (clean_df['pda'] > 0.1).astype(int)
clean_df
```
